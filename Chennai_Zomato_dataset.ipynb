__Q11) Find the list of Affordable Restaurants? ----The criteria for Affordable Restaurants would be:-----__

1)Low Price 
2) High Rated 
__First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe. Let me explain:-The most expensive restaurant has an average meal cost= 6000. Well try to stay economical and only pick the restaurants that are 1/4th of 6000.__

```python
max_price=Chennai_Zomato_df['Price'].max()
max_price
```

```python
max_price=max_price/4
max_price
```

```python
low_price = Chennai_Zomato_df[["Name of Restaurant","Location","Cuisine","Price","Top Dishes"]]
low_price=low_price[low_price['Price']<=375]
low_price.sort_values(by='Price',inplace=True)
low_price
```

```python
high_rated=Chennai_Zomato_df[["Name of Restaurant","Location","Cuisine","Top Dishes","Price","Delivery Rating","Dining Rating"]]
high_rated=high_rated[(high_rated['Delivery Rating']>4.3)&(high_rated['Dining Rating']>4.3)]
high_rated.sort_values(by=['Delivery Rating','Dining Rating'],ascending=False)
```

```python
highest_rated=pd.merge(high_rated,low_price,how='inner',on=['Name of Restaurant','Location'])
highest_rated=highest_rated.rename(columns={'Cuisine_x': 'Cuisine','Price_x': 'Price'})
highest_rated.drop(["Cuisine_y","Price_y","Top Dishes_y","Top Dishes_x"],axis=1,inplace=True)
highest_rated
```
